# 複数ニュースの取得

```python
import newspaper
url ="https://thebridge.jp/"
```

```python
website = newspaper.build(url, memorize_articles = False)
```

```python
i = 1
for article in website.articles:
    article.download()
    article.parse()

    print(f"記事{i}:{article.title}")
    print(article.url)
    print(article.text, end="\n\n")

    if i > 9:
        break

    i += 1
```

# CSVファイルへ保存

```python
article_list = []
i = 1
for article in website.articles:
    article.download()
    article.parse()

    #print(f"記事{i}:{article.title}")
    #print(article.url)
    #print(article.text, end="\n\n")

    article_list.append([i, article.title, article.url, article.text])

    if i > 9:
        break

    i += 1

print(article_list)
```

```python
import csv
import datetime

csv_header = ["記事番号", "タイトル", "URL", "本文"]
csv_date = datetime.datetime.today().strftime("%Y%m%d%H")
csv_filename = "BRIDGE_" + csv_date + ".csv"

with open(csv_filename, "w", errors="ignore") as file:
    writer = csv.writer(file, lineterminator="\n")
    writer.writerow(csv_header)
    writer.writerows(article_list)
```
