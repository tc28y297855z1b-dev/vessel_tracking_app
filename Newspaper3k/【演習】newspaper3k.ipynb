# Part2 複数の記事の取得

### タスク1: ITMedia( https://www.itmedia.co.jp/ )のサイトのトップページに表示されている記事を順に10本取得し、タイトル、URLと記事の内容を表示してください。

```python
import newspaper
url = "https://www.itmedia.co.jp/"
website = newspaper.build(url, memorize_articles=False)
```

```python
i = 0

for article in website.articles:
    article.download()
    article.parse()

    if i > 10:
        break

    print(article.title, article.url, article.text, end="\n\n")

    i += 1
```
